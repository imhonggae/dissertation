```python
import pandas as pd
```

```python
# 读取数据
df_2011 = pd.read_csv('data/lsoa-data.csv')
df_2021 = pd.read_csv('data/95-23_Median house price.csv')

# 保留 df_2011 中的指定列，并重命名
df_2011_filtered = df_2011[['Lower Super Output Area', 'House Prices;Median Price (?);2011']]
df_2011_filtered.rename(columns={'Lower Super Output Area': 'LSOA Code',
                                 'House Prices;Median Price (?);2011': '2011'}, inplace=True)
```

```python
# 保留 df_2021 中 'Local authority code' 列中以 'E09' 开头的行
df_2021_filtered = df_2021[df_2021['Local authority code'].str.startswith('E09')]

# 只保留指定的列，并重命名
df_2021_filtered = df_2021_filtered[['LSOA code', 'Year ending Dec 2021']]
df_2021_filtered.rename(columns={'LSOA code': 'LSOA Code',
                                 'Year ending Dec 2021': '2021'}, inplace=True)

# 确认过滤和重命名结果
print(df_2011_filtered.head())
print(df_2021_filtered.head())
```

```python
import pandas as pd

# 定义一个函数去除逗号并转换为数值
def clean_numeric_column(series):
    # 如果 series 不是字符串类型，先转换为字符串类型
    series = series.astype(str)
    # 去除逗号并将非数值转为 NaN
    return pd.to_numeric(series.str.replace(',', ''), errors='coerce')

# 确保 `2011` 和 `2021` 列是数值类型
df_2011_filtered['2011'] = clean_numeric_column(df_2011_filtered['2011'])
df_2021_filtered['2021'] = clean_numeric_column(df_2021_filtered['2021'])

# 合并数据，以 df_2011_filtered 为主
merged_df = pd.merge(df_2011_filtered, df_2021_filtered, on='LSOA Code', how='left')

# 处理空值，用列的中位数代替
merged_df['2011'] = merged_df['2011'].fillna(merged_df['2011'].median())
merged_df['2021'] = merged_df['2021'].fillna(merged_df['2021'].median())

# 确认合并结果和空值处理
print(merged_df)
```

```python
# 新增一列 `change`，计算 2021 和 2011 的差值
merged_df['change'] = merged_df['2021'] - merged_df['2011']

# 确认 `change` 列的计算结果
print(merged_df)
merged_df.to_csv('description_data/median_house_price.csv', index=False)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 读取shapefile
gdf = gpd.read_file('data/statistical-gis-boundaries-london/ESRI/LSOA_2011_London_gen_MHW.shp')

# 合并shapefile和处理好的数据
gdf = gdf.merge(merged_df, left_on='LSOA11CD', right_on='LSOA Code', how='left')

# 设置地图的投影
gdf = gdf.to_crs(epsg=4326)

# 创建子图
fig, axes = plt.subplots(1, 3, figsize=(20, 10))

# 绘制2011年的地图
gdf.plot(column='2011', cmap='Blues', linewidth=0.8, ax=axes[0], edgecolor='0.8', legend=True)
axes[0].set_title('2011 median house price')

# 绘制2021年的地图
gdf.plot(column='2021', cmap='Greens', linewidth=0.8, ax=axes[1], edgecolor='0.8', legend=True)
axes[1].set_title('2021 median house price')

# 绘制变化的地图
gdf.plot(column='change', cmap='RdYlBu', linewidth=0.8, ax=axes[2], edgecolor='0.8', legend=True)
axes[2].set_title('Change median house price (2021 - 2011)')

# 隐藏坐标轴
for ax in axes:
    ax.set_axis_off()

# 显示地图
plt.tight_layout()
plt.show()
```

```python
# 绘制2011年的地图
fig1, ax1 = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='2011', cmap='Blues', linewidth=0.8, ax=ax1, edgecolor='0.8', legend=True)
ax1.set_title('2011 median house price')
ax1.set_axis_off()
plt.tight_layout()
plt.show()

# 绘制2021年的地图
fig2, ax2 = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='2021', cmap='Greens', linewidth=0.8, ax=ax2, edgecolor='0.8', legend=True)
ax2.set_title('2021 median house price')
ax2.set_axis_off()
plt.tight_layout()
plt.show()

# 绘制变化的地图
fig3, ax3 = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(column='change', cmap='RdYlBu', linewidth=0.8, ax=ax3, edgecolor='0.8', legend=True)
ax3.set_title('Change median house price (2021 - 2011)')
ax3.set_axis_off()
plt.tight_layout()
plt.show()
```
