```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot

# 假设数据已经加载到一个DataFrame中
df = pd.read_csv('analysis_data/DID.csv')

# 计算变化比例
df['ethnic_change'] = (df['ethnic_2021'] - df['ethnic_2011']) / df['ethnic_2011']
df['education_change'] = (df['education_2021'] - df['education_2011']) / df['education_2011']
df['elderly_change'] = (df['elderly_2021'] - df['elderly_2011']) / df['elderly_2011']
df['unemployment_change'] = (df['unemployed_2021'] - df['unemployed_2011']) / df['unemployed_2011']
df['house_price_change'] = (df['median_house_price_2021'] - df['median_house_price_2011']) / df['median_house_price_2011']
df['crime_change_ratio'] = (df['crime_2021'] - df['crime_2011']) / df['crime_2011']

# 定义处理变量 (treatment) 和协变量 (covariates)
df['treatment'] = df['gentrification_score'] > 0
covariates = ['ethnic_change', 'education_change', 'elderly_change', 'unemployment_change', 'house_price_change']

# 分离处理组和对照组
treated = df[df['treatment'] == True]
control = df[df['treatment'] == False]

# 标准化协变量
scaler = StandardScaler()
X_treated = scaler.fit_transform(treated[covariates])
X_control = scaler.transform(control[covariates])

# 计算最近邻匹配
distances = cdist(X_treated, X_control, 'euclidean')
min_indices = np.argmin(distances, axis=1)

# 获取匹配的控制组样本
matched_control = control.iloc[min_indices]

# 合并处理组和匹配的控制组
matched_df = pd.concat([treated, matched_control])

# 计算匹配前后的标准化均值差（SMD）
def calculate_smd(treated, control, covariates):
    smd_values = {}
    for covariate in covariates:
        mean_treated = treated[covariate].mean()
        mean_control = control[covariate].mean()
        pooled_std = np.sqrt((treated[covariate].var() + control[covariate].var()) / 2)
        smd = np.abs(mean_treated - mean_control) / pooled_std
        smd_values[covariate] = smd
    return smd_values

# 计算SMD
smd_before = calculate_smd(treated, control, covariates)
smd_after = calculate_smd(matched_df[matched_df['treatment'] == True], 
                          matched_df[matched_df['treatment'] == False], covariates)

# 将SMD的结果存储在一个DataFrame中
smd_df = pd.DataFrame({'Covariate': covariates, 
                       'SMD Before Matching': list(smd_before.values()), 
                       'SMD After Matching': list(smd_after.values())})

# 显示SMD的具体数值表
print(smd_df)

# 绘制SMD变化图
smd_df.set_index('Covariate').plot(kind='barh', figsize=(10, 6))
plt.axvline(0.1, color='r', linestyle='--')
plt.title('Standardized Mean Differences (SMD) Before and After Matching')
plt.xlabel('SMD')
plt.ylabel('Covariates')
plt.show()

# 估计平均处理效应（ATT 和 ATE）
def calculate_att_ate(treated, control, outcome_var):
    # ATT: 在实际接受处理的个体中估计的处理效应
    att = treated[outcome_var].mean() - control[outcome_var].mean()
    
    # ATE: 在整体样本中估计的处理效应（处理组和控制组之间的平均差异）
    ate = df[df['treatment'] == True][outcome_var].mean() - df[df['treatment'] == False][outcome_var].mean()
    
    return att, ate

# 计算匹配后的平均处理效应（ATT 和 ATE）
att, ate = calculate_att_ate(matched_df[matched_df['treatment'] == True], 
                             matched_df[matched_df['treatment'] == False], 
                             outcome_var)

print(f"Estimated Average Treatment Effect (ATT): {att:.4f}")
print(f"Estimated Average Treatment Effect (ATE): {ate:.4f}")
```
