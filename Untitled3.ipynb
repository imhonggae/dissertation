```python
import pandas as pd
import geopandas as gpd
import numpy as np
from sklearn.neighbors import KernelDensity
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
# 加载Shapefile
shapefile_path = 'data/statistical-gis-boundaries-london/ESRI/LSOA_2011_London_gen_MHW.shp'
areas_gdf = gpd.read_file(shapefile_path)

# 加载包含绅士化得分和其他指标的CSV数据
data_path = 'analysis_data/DID.csv'
data_df = pd.read_csv(data_path)
```

```python
# 以shapefile为主表，基于 'LSOA11CD' 和 'Area' 字段进行合并
gdf = areas_gdf.merge(data_df, left_on='LSOA11CD', right_on='Area', how='left')

# 检查合并后的数据
print(gdf.head())
```

```python
gdf = gdf.dropna()

# 确保只处理Point类型的几何体，非Point类型转换为质心
gdf['geometry'] = gdf['geometry'].centroid

# 计算重心
def compute_center_of_gravity(gdf, crime_column):
    # 计算X和Y坐标的加权平均
    x_weighted_sum = (gdf.geometry.x * gdf[crime_column]).sum()
    y_weighted_sum = (gdf.geometry.y * gdf[crime_column]).sum()
    
    total_crime = gdf[crime_column].sum()
    
    x_center = x_weighted_sum / total_crime
    y_center = y_weighted_sum / total_crime
    
    return x_center, y_center

# 计算2011年和2021年的犯罪重心
center_2011 = compute_center_of_gravity(gdf, 'crime_2011')
center_2021 = compute_center_of_gravity(gdf, 'crime_2021')

print(f"2011年犯罪重心: {center_2011}")
print(f"2021年犯罪重心: {center_2021}")
```

```python
# 计算绅士化重心
def compute_gentrification_center_of_gravity(gdf):
    x_weighted_sum = (gdf.geometry.x * gdf['gentrification_score']).sum()
    y_weighted_sum = (gdf.geometry.y * gdf['gentrification_score']).sum()
    
    total_gentrification = gdf['gentrification_score'].sum()
    
    x_center = x_weighted_sum / total_gentrification
    y_center = y_weighted_sum / total_gentrification
    
    return x_center, y_center

# 计算绅士化重心
center_gentrification = compute_gentrification_center_of_gravity(gdf)

print(f"绅士化重心: {center_gentrification}")
```

```python
import contextily as ctx
import matplotlib.pyplot as plt

# 绘制犯罪重心和绅士化重心
fig, ax = plt.subplots(figsize=(10, 10))

# 将数据转换为Web Mercator坐标系，以便正确加载底图
gdf = gdf.to_crs(epsg=3857)

# 绘制地图的边界框
gdf.plot(ax=ax, alpha=0)  # 不显示多边形，只用于设置边界框

# 手动设置地图的显示范围 (基于数据的范围)
xlim = ([gdf.total_bounds[0], gdf.total_bounds[2]])
ylim = ([gdf.total_bounds[1], gdf.total_bounds[3]])
ax.set_xlim(xlim)
ax.set_ylim(ylim)

# 使用OpenStreetMap作为底图，并设置缩放级别
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik, zoom=12)

# 确保重心点的坐标系与地图一致
center_2011 = gdf.to_crs(epsg=3857).geometry.centroid.iloc[0].x, gdf.to_crs(epsg=3857).geometry.centroid.iloc[0].y
center_2021 = gdf.to_crs(epsg=3857).geometry.centroid.iloc[1].x, gdf.to_crs(epsg=3857).geometry.centroid.iloc[1].y
center_gentrification = gdf.to_crs(epsg=3857).geometry.centroid.iloc[2].x, gdf.to_crs(epsg=3857).geometry.centroid.iloc[2].y

# 绘制重心点，确保它们在底图之上显示，并增加点的大小
ax.scatter(*center_2011, color='red', label='2011 Crime Center', zorder=5, s=100)
ax.scatter(*center_2021, color='blue', label='2021 Crime Center', zorder=5, s=100)
ax.scatter(*center_gentrification, color='green', label='Gentrification Center', zorder=5, s=100)

# 添加图例
plt.legend()

# 显示图像
plt.show()

```

```python
import pandas as pd
import geopandas as gpd
from scipy.stats import ttest_ind

# 计算犯罪变化率
gdf['crime_change_ratio'] = (gdf['crime_2021'] - gdf['crime_2011']) / gdf['crime_2011']

# 将区域分类为绅士化和非绅士化
gentrified_areas = gdf[gdf['gentrification_score'] > 0]
non_gentrified_areas = gdf[gdf['gentrification_score'] <= 0]

# 计算两个类别的平均犯罪变化率
mean_crime_change_gentrified = gentrified_areas['crime_change_ratio'].mean()
mean_crime_change_non_gentrified = non_gentrified_areas['crime_change_ratio'].mean()

print(f"绅士化区域的平均犯罪变化率: {mean_crime_change_gentrified}")
print(f"非绅士化区域的平均犯罪变化率: {mean_crime_change_non_gentrified}")

# 使用t检验比较绅士化区域和非绅士化区域的犯罪变化
t_stat, p_value = ttest_ind(gentrified_areas['crime_change_ratio'], non_gentrified_areas['crime_change_ratio'])

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# 判断是否存在显著差异
if p_value < 0.05:
    print("绅士化区域与非绅士化区域的犯罪变化率存在显著差异。")
else:
    print("绅士化区域与非绅士化区域的犯罪变化率没有显著差异。")
```
