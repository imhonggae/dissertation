```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
path = 'analysis_data/indicator.csv'
df = pd.read_csv(path)
```

```python
# 计算每个指标的变化比例
df['ethnic_score'] = (df['ethnic_2021'] - df['ethnic_2011']) / df['ethnic_2021']
df['education_score'] = (df['education_2021'] - df['education_2011']) / df['education_2021']
df['elderly_score'] = (df['elderly_2021'] - df['elderly_2011']) / df['elderly_2021']
df['unemployed_score'] = (df['unemployed_2021'] - df['unemployed_2011']) / df['unemployed_2021']
df['median_house_price_score'] = (df['median_house_price_2021'] - df['median_house_price_2011']) / df['median_house_price_2021']

# 计算中产阶级化得分
df['gentrification_score'] = (df['education_score'] +
                              df['median_house_price_score'] -
                              df['elderly_score'] -
                              df['ethnic_score'] -
                              df['unemployed_score']
                              )

score = df[['Area', 'ethnic_score', 'education_score', 'elderly_score', 'unemployed_score', 'median_house_price_score', 'gentrification_score']]

# 显示得分
print(score)
```

```python
# 读取.shp文件
shapefile_path = 'data/statistical-gis-boundaries-london/ESRI/LSOA_2011_London_gen_MHW.shp'
gdf = gpd.read_file(shapefile_path)

# 合并数据
gdf = gdf.merge(score, left_on='LSOA11CD', right_on='Area')

# 创建地图图层
fig, ax = plt.subplots(1, 1, figsize=(15, 15))

# 绘制非特定区域的地图
gdf[gdf['LAD11CD'] != 'E09000001'].plot(column='gentrification_score', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# 绘制特定区域的地图，用绿色表示
gdf[gdf['LAD11CD'] == 'E09000001'].plot(ax=ax, color='green', linewidth=0.8, edgecolor='0.8')

# 设置地图标题
ax.set_title('Gentrification Score in London (2011-2021)', fontsize=20)

# 显示地图
plt.show()

```
